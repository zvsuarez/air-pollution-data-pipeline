```python
import pandas as pd
```

```python
df = pd.read_csv('worldcities.csv').convert_dtypes()
df
```

```python
city = df[['city_ascii', 'country','lat', 'lng', 'population']]
city.loc[:, 'country'] = city['country'].replace('Congo (Kinshasa)', 'Congo')
city = city.drop_duplicates('country', keep='first').sort_values('population', ascending=False)[:60].reset_index(drop=bool)
#city.to_csv('coordinates.csv')
```
